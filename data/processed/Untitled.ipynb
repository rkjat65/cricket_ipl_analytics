```python
import pandas as pd
```

```python
df = pd.read_csv('ipl_2021_matches.csv')
```

```python
df_clean = df.drop_duplicates(subset=['match_date', 'team1', 'team2', 'venue'], keep='first')
df_clean.to_csv('ipl_2021_matches.csv', index=False)
print(f"2021: {len(df)} → {len(df_clean)} matches")
```

```python
# Quick diagnostic
df=pd.read_csv('ipl_2021_matches.csv'); 
print(df[['match_date', 'team1', 'team2', 'venue']].drop_duplicates().to_string()); 
print(f'\n\nUnique dates: {df.match_date.nunique()}, Unique venues: {df.venue.nunique()}')
```
